# **Ejercicio 1: Normalización Min-Max**

**1. Cargar el dataset wine de sklearn.**

```python
from IPython.display import display, HTML
display(HTML("<h3 style='color:blue; font-size:18px;'> Dataset de wine </h3>"))

import pandas as pd
from sklearn.datasets import load_wine

df = load_wine()
dfwine = pd.DataFrame(df.data, columns=df.feature_names)
print(dfwine)
```

**2. Seleccionar una de las características numéricas (malic_acid).**

```python
display(HTML("<h3 style='color:blue; font-size:18px;'> Se eligió la columna malic_acid: se muestran los datos descriptivos de esta variable </h3>"))
print('\n')
print(dfwine['malic_acid'].describe())
```

**3. Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.**

```python
display(HTML("<h3 style='color:blue; font-size:18px;'> Pruebas de normalidad </h3>"))
print('\n')

import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

x = dfwine['malic_acid']


# Pruebas de normalidad
print("Pruebas de normalidad para 'malic_acid' ")
print(f"Media: {x.mean():.3f}, Desviación estándar: {x.std():.3f}")

# Shapiro–Wilk
shapiro_test = stats.shapiro(x)
print(f"Shapiro–Wilk: estadístico = {shapiro_test.statistic:.5f}, p-valor = {shapiro_test.pvalue:.5f}")

# D’Agostino–Pearson
dagostino_test = stats.normaltest(x)
print(f"D’Agostino–Pearson: estadístico = {dagostino_test.statistic:.5f}, p-valor = {dagostino_test.pvalue:.5f}")

# Kolmogorov–Smirnov (con parámetros estimados)
z = (x - x.mean()) / x.std(ddof=1)
ks_test = stats.kstest(z, 'norm')
print(f"Kolmogorov–Smirnov: estadístico = {ks_test.statistic:.5f}, p-valor = {ks_test.pvalue:.5f}")

print('\n')


# Visualización
plt.figure(figsize=(12, 5))

# Histograma + KDE
plt.subplot(1, 2, 1)
sns.histplot(x, bins=20, kde=True, color='skyblue')
plt.title("Distribución de malic_acid")
plt.xlabel("malic_acid")
plt.ylabel("Frecuencia")

# QQ-Plot
plt.subplot(1, 2, 2)
stats.probplot(x, dist="norm", plot=plt)
plt.title("QQ-Plot de malic_acid")

plt.tight_layout()
plt.show()
```

**4. Aplicar la normalización Min-Max a los datos seleccionados.**

```python
display(HTML("<h3 style='color:blue; font-size:18px;'> Normalización por el método Min-Max </h3>"))
print('\n')

from sklearn.preprocessing import MinMaxScaler

# Crear y aplicar el escalador con el rango [0,1]
scaler = MinMaxScaler(feature_range=(0, 1))
X_malic_acid = scaler.fit_transform(dfwine[['malic_acid']])

X_normalizada = pd.Series(X_malic_acid.ravel(), index=dfwine.index, name='X_normalizada')

print("Datos originales:\n", dfwine['malic_acid'].head())
print('\n')
print("Datos normalizados:\n", X_normalizada.head())
```

**5. Realizar la prueba de normalidad nuevamente sobre los datos normalizados.**

```python
display(HTML("<h3 style='color:blue; font-size:18px;'> Pruebas de normalidad de los datos normalizados </h3>"))
print('\n')

x2 = X_normalizada

# Pruebas de normalidad
print("Pruebas de normalidad para 'malic_acid (normalizada: min-max)' ")
print(f"Media: {x2.mean():.3f}, Desviación estándar: {x2.std():.3f}")

# Shapiro–Wilk
shapiro_test = stats.shapiro(x2)
print(f"Shapiro–Wilk: estadístico = {shapiro_test.statistic:.5f}, p-valor = {shapiro_test.pvalue:.5f}")

# D’Agostino–Pearson
dagostino_test = stats.normaltest(x2)
print(f"D’Agostino–Pearson: estadístico = {dagostino_test.statistic:.5f}, p-valor = {dagostino_test.pvalue:.5f}")

# Kolmogorov–Smirnov (con parámetros estimados)
z = (x2 - x2.mean()) / x2.std(ddof=1)
ks_test = stats.kstest(z, 'norm')
print(f"Kolmogorov–Smirnov: estadístico = {ks_test.statistic:.5f}, p-valor = {ks_test.pvalue:.5f}")

print('\n')


# Visualización
plt.figure(figsize=(12, 5))

# Histograma + KDE
plt.subplot(1, 2, 1)
sns.histplot(x2, bins=20, kde=True, color='skyblue')
plt.title("Distribución de malic_acid malic_acid (normalizada: min-max)")
plt.xlabel("malic_acid")
plt.ylabel("Frecuencia")

# QQ-Plot
plt.subplot(1, 2, 2)
stats.probplot(x2, dist="norm", plot=plt)
plt.title("QQ-Plot de malic_acid malic_acid (normalizada: min-max)")

plt.tight_layout()
plt.show()
```

**6. Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.**

```python
import matplotlib.pyplot as plt
import seaborn as sns

x = dfwine['malic_acid']
x2 = X_normalizada

# Visualización
plt.figure(figsize=(12, 5))

# Histograma + KDE para datos originales
plt.subplot(1, 2, 1)
sns.histplot(x, bins=20, kde=True, color='skyblue')
plt.title("Distribución de malic_acid (Original)")
plt.xlabel("malic_acid")
plt.ylabel("Frecuencia")

# Histograma + KDE para datos normalizados
plt.subplot(1, 2, 2)
sns.histplot(x2, bins=20, kde=True, color='salmon') # Using a different color for clarity
plt.title("Distribución de malic_acid (Normalizada Min-Max)")
plt.xlabel("malic_acid")
plt.ylabel("Frecuencia")


plt.tight_layout()
plt.show()
```

**¿Cómo afectó la normalización Min-Max a la distribución de los datos?**

Para el caso de la variable 'malic_acid' no se percibe ningún cambio en la distribución de los datos, ya que, la normalización Min-Max no modifica la forma de la distribución; solo reescala los valores a un nuevo rango, este caso [0,1].

**¿Qué interpretas del p-valor antes y después de la normalización?**

Al igual que la distribución de los datos, en el caso del valor (p), tampoco cambio la probabilidad de que dichos datos normalizados se ajuste una distribución normal. Se puede concluir que este método no es útil para datos que presentan una distribución sesgada hacia la derecha.

**1. Cargar un conjunto de datos aleatorios generados con numpy.**

```python
import numpy as np
#np.random.seed(42)

# Distribución Weibull
shape = 7.0
scale = 15.0
n = 1000

# Generar números aleatorios Weibull
x_weibull = np.random.weibull(a=shape, size=n) * scale  # se multiplica por scale para ajustar

# Mostrar los primeros valores
print(x_weibull[:100])
```

**2. Aplicar una transformación logarítmica (np.log) a los datos.**

```python
display(HTML("<h3 style='color:blue; font-size:18px;'> Normalización por el método de Transformación Logarítmica </h3>"))
print('\n')


x_weibull = pd.Series(x_weibull, name='x_weibull')
x_log = np.log(x_weibull)
x_estandarizada = pd.Series(x_log.ravel(), index=x_weibull.index, name='x_estandarizada')

print("Datos originales:\n", x_weibull.head())
print('\n')
print("Datos estandarizados:\n", x_estandarizada.head())
```

**3. Realizar una prueba de normalidad sobre los datos originales y transformados.**

# **Ejercicio 4: Transformación Raíz Cuadrada**


**1. Cargar un conjunto de datos aleatorios generados con numpy.**

```python
import numpy as np
#np.random.seed(42)

# Distribución Weibull
shape = 7.0
scale = 15.0
n = 1000

# Generar números aleatorios Weibull
x_weibull = np.random.weibull(a=shape, size=n) * scale  # se multiplica por scale para ajustar

# Mostrar los primeros valores
print(x_weibull[:100])
```

**2. Aplicar una transformación de raíz cuadrada a los datos.**
