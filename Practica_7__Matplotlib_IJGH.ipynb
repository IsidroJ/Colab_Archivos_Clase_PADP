# **Practica 7. Análisis de un Dataset con Matplotlib**

Utilizar matplotlib para crear gráficos visuales que permitan analizar e interpretar los datos.

1. Carga y exploración del dataset
2. Gráfico de barras
3. Histograma
4. Gráfico de dispersión (scatter plot)
5. Gráfico de cajas (box plot)
6. Gráfico de líneas





**1. Carga y exploración del dataset**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

dataframe = sns.load_dataset('iris')
print("Las primeras filas del conjunto de datos de ('iris'): \n", dataframe.head())

print('\n')

print("Información general del conjunto de datos de ('iris'): \n")
print(dataframe.info())
```

**2. Gráfico de barras**

```python
PromedioLyA = dataframe.groupby('species')[['petal_length', 'petal_width']].mean()

PromedioLyA.plot(kind='bar', color=['#0000FF', '#FF0000'], figsize=(8,5))  # figsize=(9,6) define el tamaño de la figura
plt.title('Promedio del largo y ancho del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Promedio')
plt.xticks(rotation=0)
plt.legend(['Largo del pétalo', 'Ancho del pétalo'], title="Medidas")
plt.show()
```

**3. Histograma**

```python
import numpy as np

n = 150
k = 1 + 3.322 * np.log10(n)

plt.figure(figsize=(7,5)) #para que se vea reflejado el tamo de l afigura la instrucción debe ir antes de mandar a llamar al histograma
plt.hist(dataframe['sepal_length'], bins=round(k), edgecolor='black', color='#FF7F50')
plt.title('Histograma de frecuencias de las longitudes del sépalo')
plt.xlabel('Longitud del Sépalo')
plt.ylabel('Frecuencia')
plt.show()
```

**4. Gráfico de dispersión**

```python
for especie in dataframe['species'].unique():
    subset = dataframe[dataframe['species'] == especie]
    plt.scatter(subset['petal_length'], subset['petal_width'], label=especie)

plt.title('Relación del largo y ancho del pétalo por especie')
plt.xlabel('Largo del pétalo')
plt.ylabel('Ancho del pétalo')
plt.legend(title="Especies")
plt.grid(True)
plt.show()
```

**5. Gráfico de cajas**

```python
plt.figure(figsize=(10,6))
dataframe.boxplot(column=['sepal_length'],
                  by='species',
                  patch_artist=True,
                  boxprops=dict(facecolor="#5F9EA0", color="#FF7F50"),
                  medianprops=dict(color="orange", linewidth=2),)
plt.suptitle('') # Quitar título automático
plt.title('Distribución del largo del sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Largo')
plt.show()
```

```python
plt.figure(figsize=(10,6))
dataframe.boxplot(column=['sepal_width'], by='species')
plt.suptitle('') # Quitar título automático
plt.title('Distribución del ancho del sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Ancho')
plt.show()
```

**6. Gráfico de líneas**

```python
plt.plot(dataframe.index, dataframe['petal_length'], linestyle='-', marker='', color='#8A2BE2')
plt.title('Tendencia de la longitud del pétalo')
plt.xlabel('Observación i')
plt.ylabel('Longitud del pétalo')
plt.show()
```

```python
plt.figure(figsize=(10,5))

for especie in dataframe['species'].unique():
    subset = dataframe[dataframe['species'] == especie]
    plt.plot(subset.index, subset['petal_length'], label=especie)

plt.title('Tendencia de la longitud del pétalo')
plt.xlabel('Observación i')
plt.ylabel('Longitud del pétalo')
plt.legend(title="Especies")
plt.show()
```
