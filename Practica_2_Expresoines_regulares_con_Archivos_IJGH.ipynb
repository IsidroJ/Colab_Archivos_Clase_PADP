**Importancia de la Expresiones Regulares "regex"**

Las expresiones regulares (regex) son una herramienta fundamental en el análisis y procesamiento de datos, en específico para trabajar con texto, ya que permiten buscar, identificar y extraer patrones específicos de manera eficiente. El uso de regex resulta esencial para filtrar información clave contenida en un archivo, el cual puede contener símbolos, números, fechas, texto, etc. Mediante el empleo de regex, es posible automatizar la detección de estos elementos sin necesidad de revisar manualmente grandes volúmenes de datos o información, lo que incrementa la velocidad y precisión del análisis. Además, esta técnica es ampliamente utilizada en áreas como minería de datos, análisis financiero, validación de entradas y limpieza de información. Por ello, aprender a construir y aplicar expresiones regulares no solo optimiza tareas de extracción de datos, sino que también desarrolla habilidades valiosas para el tratamiento avanzado de información en entornos profesionales y académicos.

**Algunas de las expresiones que se pueden utilziar para la extración de información son las siguientes:**


> ^ Coincide con el comienzo de la línea.

> $ Coincide con el final de la línea

> . Coincide con cualquier carácter (un comodín).

> \s Coincide con un espacio en blanco.

> \S Coincide con un carácter que no sea un espacio en blanco (el opuesto a \s).

> * Se aplica al carácter o caracteres inmediatamente anteriores, indicando que pueden coincidir cero o más veces.

> *? Se aplica al carácter o caracteres inmediatamente anteriores, indicando que coinciden cero o más veces en modo “no ambicioso”.

> + Se aplica al carácter o caracteres inmediatamente anteriores, indicando que pueden coincidir una o más veces.

> +? Se aplica al carácter o caracteres inmediatamente anteriores, indicando que pueden coincidir una o más veces en modo “no ambicioso”.

> ? Se aplica al carácter o caracteres inmediatamente anteriores, indicando que puede coincidir cero o una vez.

> ?? Se aplica al carácter o caracteres inmediatamente anteriores, indicando que puede coincidir cero o una vez en modo “no ambicioso”.

> [aeiou] Coincide con un solo carácter, siempre que éste se encuentre dentro del conjunto especificado. En este caso, coincidiría con “a”, “e”, “i”, “o”, o “u”, pero no con otros caracteres.

> [a-z0-9] Se pueden especificar rangos de caracteres utilizando el signo menos. En este caso, sería un solo carácter que debe ser una letra minúscula o un dígito.

> [^A-Za-z] Cuando el primer carácter en la notación del conjunto es “^”, invierte la lógica. En este ejemplo, habría coincidencia con un solo carácter que no sea una letra mayúscula o una letra minúscula.

> ( ) Cuando se agregan paréntesis a una expresión regular, son ignorados para propósitos de encontrar coincidencias, pero permiten extraer un subconjunto determinado de la cadena en que se encuentra la coincidencia, en lugar de toda la cadena como cuando se utiliza findall().

> \b Coincide con una cadena vacía, pero solo al comienzo o al final de una palabra.

> \B Concide con una cadena vacía, pero no al comienzo o al final de una palabra.

> \d Coincide con cualquier dígito decimal; equivalente al conjunto [0-9].

> \D Coincide con cualquier carácter que no sea un dígito; equivalente al conjunto [^0-9].

# 4.1 El siguiente código extrae la información (correos) de las personas que enviaron un mail.

```python
import re
man = open('/content/drive/MyDrive/1_ClassFiles/informe_acciones.txt', encoding='latin-1')
for linea in man:
    linea = linea.rstrip()
    if re.search('De:', linea):
        print(linea)
```

# 4.3 El siguiente código extrae el nombre de la empresa


*   re.search(...) localiza la parte del texto después de "valor actual de la acción de" y antes del paréntesis con el ticker.

*   (.+?) es un grupo de captura que guarda el nombre completo de la empresa.

*   re.IGNORECASE hace que la búsqueda no dependa de mayúsculas/minúsculas.

```python
import re
man = open('/content/drive/MyDrive/1_ClassFiles/informe_acciones.txt', encoding='latin-1')
for linea in man:
    linea = linea.rstrip()
    coincidir = re.search(r"Te escribo para informarte sobre el valor actual de la acción de\s+(.+?)\s+\(", linea, re.IGNORECASE)
    if coincidir:
        empresa = coincidir.group(1)
        print(empresa)
```

# 4.4 El siguiente código extrae el nombre de la empresa y su acrónimo


*   valor actual de la acción de\s+ → busca esa frase y los espacios que le siguen.

*   De la siguinete expresión (.+\([A-Z.]+\)) →

*   ".+" → captura cualquier texto (nombre de la empresa).

*   "\([A-Z.]+\)" → busca un paréntesis con letras mayúsculas y/o puntos.

*   Los paréntesis () en el patrón indican grupo de captura, por eso luego usamos match.group(1) para obtener solo esa parte.

```python
import re
man = open('/content/drive/MyDrive/1_ClassFiles/informe_acciones.txt', encoding='latin-1')
for linea in man:
    linea = linea.rstrip()
    coincidir = re.search(r"Te escribo para informarte sobre el valor actual de la acción de\s+(.+\([A-Z.]+\))", linea, re.IGNORECASE)
    if coincidir:
        empresa = coincidir.group(1)
        print(empresa)
```

#4.5 El siguiente código extrae el acronimo o siglas de la empresa

*   "\(" y "\)" buscan paréntesis literales.

*   "[A-Z.]+" uno o más caracteres en mayúsculas o puntos.

*   "coincidir.group()" devuelve el texto que coincide, incluyendo los paréntesis.


```python
import re
man = open('/content/drive/MyDrive/1_ClassFiles/informe_acciones.txt', encoding='latin-1')
for linea in man:
    linea = linea.rstrip()
    coincidir = re.search(r"\([A-Z.]+\)", linea)
    if coincidir:
        print(coincidir.group())

```
